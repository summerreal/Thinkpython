```python
#字典作为计数器集合,计算一个字符串每个字母出现的次数。11章
def histogram(s):
    d=dict()
    for c in s:
        d[c] = d.get(c,0) + 1 #接受一个键和一个默认值作为参数。 如果字典中存在该键，则返回对应值；否则返回传入的0值。
    return d

h = histogram('brontosaurus')
h
```

```python
#接受一个字符串，并按字符出现频率降序打印字母
def most_frequent(s):
    d= histogram(s)
    t = list()
    for key,value in d.items(): #items返回由多个元组组成的序列，其中每个元组是一个键值对。
        t.append((value,key))
    t.sort()
    t.reverse() #反向列表中元素。
    return t

most_frequent('brontosaurus')
    
```

如果两个单词中的某一单词可以通过调换两个字母变为另一个，这两个单词就构成了“换位对（metatheisi pair）”；比如，“converse”和“conserve”。 编写一个程序，找出字典里所有的“换位对”。

提示：不用测试所有的单词组合，也不用测试所有的字母调换组合。致谢：这道习题受 http://puzzlers.org 上的案例启发而来

```python
#“换位对”函数判断
def ispair(s1,s2):
    if sorted(s1) ==sorted(s2):#比较两个单词字母一致
        count =0
        for x,y in zip(s1,s2): #调换两个字母
            if x == y:
                count += 1

        if count == len(s1)-2:
            return True
    return False


#打开文件
fin =open(r'C:\Users\Administrator\Desktop\summer\exercise\chap12\words.txt')
d = {}
for line in fin:
    word =line.strip() #字符串方法 strip 删掉空格
    d[word] = len(word) #读取文件写入字典,构建单词长度的字典

for i in d:
    for ii in d:
        if d[i] == d[ii]: #单词长度一样
            if ispair(i,ii):#调用函数比较两个单词字母一致，是否是换位对
                print(i,ii)

```
