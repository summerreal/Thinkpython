```python
#11-2
# #反转字典
def invert_dict(d):
    inverse = {}
    for key in d:
        temp = d[key]
        inverse.setdefault(temp,[]).append(key)
    return inverse


d={'a':1,'b':2,'c':3}
print(d)
inverse=invert_dict(d)
print(inverse)
for key in inverse:
    print(key,inverse[key])
```

将习题6-2中的Ackermann函数备忘录化（memoize），看看备忘录化（memoization）是否可以支持解决更大的参数。没有提示！


```python
# 11-3
#ackerman函数实现，递归
know = {}

def ack(m:int,n:int):
   
    if m==0:
        return n+1
    if n==0 and m>0:
        return ack(m-1,1)
    try:
        return know[m,n] #检查值是否已经存在，如果是直接返回值
    except KeyError:
        know[m,n] = ack(m-1,ack(m,n-1)) #记录已经计算过的阿克曼函数值
        return know[m,n]

print(ack(3,4))
print(ack(3,5))
print(ack(1,3))
```

习题11-4

如果你做了习题10-7，你就已经写过一个叫 has_duplicates 的函数，它接受一个列表作为参数，如果其中有某个对象在列表中出现不止一次就返回True。

用字典写个更快、更简单的版本。

```python
#10-7程序回顾
def has_duplicates(t):
    t=sorted(t)
    for i in range(len(t)-1):
        if t[i]==t[i+1]:
            return True

t=[1,2,5,7,4,3,46,567,456,456]
print(has_duplicates(t))
```

```python
#11-4
def has_duplicates(t):
    d=dict()
    for c in t:
        if c in d:
            return True
        d[c] = 1
    return False


t= [1,2,5,7,4,3,46,567,456,456]
print(has_duplicates(t))

#运行结果可以看出，使用字典后时间节省3/7
```

两个单词如果反转其中一个就会得到另一个，则被称作“反转对”（参见习题8-5中的 rotate_word ）。

编写一程序，读入单词表并找到所有反转对。

```python
#凯撒密码

def rotate_word(s,n): #反转函数
    res=''
    for i in s:
        if s.islower():  #s都是小写
            start = ord('a')
        elif s.isupper(): 
            start =ord('A') #s都是大写
        else:
            start=i

        m=(ord(i)-start+n)%26+start #ord字符-->数值 编码
        res +=chr(m) ##ord数值-->字符转
        
    return res


def one_append(): #读取word，存入列表t
    t=[]
    fin =open(r'C:\Users\Administrator\Desktop\summer\exercise\chap11\words.txt')
    for line in fin:
        #print(line)
        word =line.strip()
        t.append(word)
    return t

t=one_append()

for i in range(len(t)): #遍历所有单词
    for j in range(len(t)):
        for n in range(14):
            if rotate_word(t[i],n) ==t[j]:
                print (t[i],t[j],'\n')

```
